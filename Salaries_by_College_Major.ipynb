```python
#IMPORTING LIBRARIES AND DEPENDANCIES
import pandas as pd
import numpy as np
```

```python
#LOADING AND READING THE DATASET
df = pd.read_csv("/content/salaries_by_college_major (1).csv")
df.head()
```

```python
#FINDING THE SHAPE OF THE DATASET
df.shape
colums = 6
rows = 51
```

```python
#FINDING THE COLUMN NAMES
column_name = df.columns
print(f"Name of columns :{column_name} ")
```

```python
df.isna()
clean_df = df.dropna()
```

```python
#COLLEGE MAJOR WITH HIGHEST STARTING SALARIES
starting_Salary = clean_df['Starting Median Salary']
max = starting_Salary.max()
print(f"Highest starting salary = {max}")

#which major?
max_major_index = starting_Salary.idxmax()
max_major = clean_df['Undergraduate Major'].loc[43]
print(f"Major with the highest starting salary = {max_major}")
```

```python
#COLLEGE MAJOR WITH THE HIGHEST MID CAREER SALARY
mid_car_salary = clean_df['Mid-Career Median Salary']
mid_max = mid_car_salary.max()
print(f"Highest mid career salary is {mid_max}")

#Which major?
mid_max_index= mid_car_salary.idxmax()
mid_max_major = clean_df['Undergraduate Major'].loc[mid_max_index]
print(f"Major with the highest mid career salary = {mid_max_major}")
```

```python
#Which college major has the lowest starting salary and how much do graduates 
#earn after university?
low_start = starting_Salary.min()
print(f"Lowest starting salary is {low_start}")

#which degree?
low_start_index =  starting_Salary.idxmin()
low_start_major = clean_df['Undergraduate Major'].loc[low_start_index]
print(f"College major with the lowest starting salary is {low_start_major}")
low_Start_mid_career = mid_car_salary.loc[low_start_index]
print(f"Students with {low_start_major} tend to earn {low_Start_mid_career} in their mid career")
```

```python
#Which college major has the lowest mid-career salary and how much can people
# expect to earn with this degree? 
low_mid = mid_car_salary.min()
print(f"Lowest mid career salary is {low_mid}")
low_mid_index = mid_car_salary.idxmin()
low_mid_major = clean_df['Undergraduate Major'].loc[low_mid_index]
print(f"Major with the lowest mid career salary is {low_mid_major}")
```

```python
spread_col = clean_df['Mid-Career 90th Percentile Salary']-clean_df['Mid-Career 10th Percentile Salary']
clean_df.insert(1, "Spread", spread_col)
```

```python
clean_df.head()
```

```python
low_risk = clean_df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
# top 5 degrees with the highest values in the 90th percentile. 
highest_90 = clean_df.sort_values('Mid-Career 90th Percentile Salary')
print("the top 5 degrees with the highest values in the 90th percentile:")
highest_90[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].tail()
```

```python
# the degrees with the greatest spread in salaries
high_risk = clean_df.sort_values('Spread')
print("the degrees with the greatest spread in salaries")
high_risk[['Undergraduate Major', 'Spread']].tail()
```

```python
#MEAN SALARIES BY GROUPS -(BUSINESS, HASS, STEM)
pd.options.display.float_format = '{:,.2f}'.format 
clean_df.groupby('Group').mean()
```
